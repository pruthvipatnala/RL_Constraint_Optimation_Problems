### Cybenko Equations

Printing equations using the coeficients obtained in cybenko_approx.ipynb

```python
import numpy as np
import math

def sigmoid(x): 
    return 1 / (1 + np.exp(-x))
```

```python
def calc_iter(episode,n):
    weights = nn_parameters[n][0]
    biases = nn_parameters[n][1]
    
    ip_h_weights = weights[0][0]
    h_op_weights = weights[1]
    
    h_biases = biases[0]
    op_bias = biases[1][0]
    
    #print(h_op_weights)
    
    w1x = [w*episode for w in ip_h_weights]
    w1x_b1 = [w1x[i] + h_biases[i] for i in range(len(h_biases))]
    
    sig = [sigmoid(i) for i in w1x_b1]
    
    w2h = [h_op_weights[i] * sig[i] for i in range(len(sig))]
    iterations = sum(w2h) + op_bias
    
    #print(iterations)
    return iterations[0]
    
    
def print_equation(n):
    weights = nn_parameters[n][0]
    biases = nn_parameters[n][1]
    
    ip_h_weights = weights[0][0]
    h_op_weights = weights[1]
    
    h_biases = biases[0]
    op_bias = biases[1][0]
    
    equation = ["sig("+str(ip_h_weights[i])+" * x + "+str(h_biases[i])+")" for i in range(len(ip_h_weights))]
    equation = [str(h_op_weights[i][0])+" * "+equation[i] for i in range(len(equation))]
    equation = ' + '.join(equation)
    if('-' in str(op_bias)):
        equation = equation +" - "+ str(op_bias)[1:]
    else:
        equation = equation +" + "+ str(op_bias)
    
    return equation
    
```
